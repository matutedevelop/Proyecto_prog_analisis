```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cuartos_disponibles = pd.read_excel('cleaned_edu62.xlsx', sheet_name='Hoja1',)
cuartos_disponibles['Fecha'] = pd.to_datetime(cuartos_disponibles['Año'].astype(str) + '-' + cuartos_disponibles['Mes'].str.extract(r'\[(\d+)\]')[0])
cuartos_disponibles
```

```python
plt.figure(figsize=(12, 6))
plt.plot(cuartos_disponibles['Fecha'], cuartos_disponibles['Cuartos_disponibles'], label='Cuartos Disponibles')
plt.xlabel('Fecha')
plt.ylabel('Cuartos Disponibles')
plt.title('Serie de Tiempo: Cuartos Disponibles')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Crear una figura con subplots
fig, axes = plt.subplots(5, 1, figsize=(12, 25), sharex=True)

# Cuartos Disponibles
axes[0].plot(cuartos_disponibles['Fecha'], cuartos_disponibles['Cuartos_disponibles'], label='Cuartos Disponibles', color='blue')
axes[0].set_title('Serie de Tiempo: Cuartos Disponibles')
axes[0].set_ylabel('Cuartos Disponibles')
axes[0].legend()
axes[0].grid(True)

# Cuartos Ocupados
axes[1].plot(Cuartos_ocupados['Fecha'], Cuartos_ocupados['Cuartos_ocupados'], label='Cuartos Ocupados', color='orange')
axes[1].set_title('Serie de Tiempo: Cuartos Ocupados')
axes[1].set_ylabel('Cuartos Ocupados')
axes[1].legend()
axes[1].grid(True)

# Llegada de Turistas
axes[2].plot(Llegada_turistas['Fecha'], Llegada_turistas['Llegada_turistas'], label='Llegada de Turistas', color='green')
axes[2].set_title('Serie de Tiempo: Llegada de Turistas')
axes[2].set_ylabel('Llegada de Turistas')
axes[2].legend()
axes[2].grid(True)

# Turistas por Noche
axes[3].plot(Turistas_noche['Fecha'], Turistas_noche['Turistas_noche'], label='Turistas por Noche', color='purple')
axes[3].set_title('Serie de Tiempo: Turistas por Noche')
axes[3].set_ylabel('Turistas por Noche')
axes[3].legend()
axes[3].grid(True)

# Porcentaje de Ocupación
axes[4].plot(Porcentage_ocupacion['Fecha'], Porcentage_ocupacion['Porcentage_ocupacion'], label='Porcentaje de Ocupación', color='red')
axes[4].set_title('Serie de Tiempo: Porcentaje de Ocupación')
axes[4].set_xlabel('Fecha')
axes[4].set_ylabel('Porcentaje de Ocupación')
axes[4].legend()
axes[4].grid(True)

# Ajustar el layout
plt.tight_layout()
plt.show()
```
