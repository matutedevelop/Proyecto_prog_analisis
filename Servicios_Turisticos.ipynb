```python
import pandas as pd

df_2013 = pd.read_csv('data\Servicios_Turisticos_Complementarios\Servicios_Turisticos_2013.csv')
df_2014 = pd.read_csv('data\Servicios_Turisticos_Complementarios\Servicios_Turisticos_2014.csv')
df_2015 = pd.read_csv('data\Servicios_Turisticos_Complementarios\Servicios_Turisticos_2015.csv')
df_2016 = pd.read_csv('data\Servicios_Turisticos_Complementarios\Servicios_Turisticos_2016.csv')
df_2017 = pd.read_csv('data\Servicios_Turisticos_Complementarios\Servicios_Turisticos_2017.csv')
df_2018 = pd.read_csv('data\Servicios_Turisticos_Complementarios\Servicios_Turisticos_2018.csv')
df_2019 = pd.read_csv('data\Servicios_Turisticos_Complementarios\Servicios_Turisticos_2019.csv')
df_2020 = pd.read_csv('data\Servicios_Turisticos_Complementarios\Servicios_Turisticos_2020.csv')

```

```python
df = pd.concat([df_2013, df_2014, df_2015, df_2016, df_2017, df_2018, df_2019, df_2020])
df 
```

```python
df = df.drop(columns=["Unnamed: 0"])
```

```python
df['municipio'] = df['municipio'].str.upper()
```

```python
df = df[df['municipio'] != 'ESTADO']
df = df[df['municipio'] != 'I NORTE']
df = df[df['municipio'] != 'II ALTOS NORTE']
df = df[df['municipio'] != 'III ALTOS SUR']
df = df[df['municipio'] != 'IV CIÉNEGA']
df = df[df['municipio'] != 'V SURESTE']
df = df[df['municipio'] != 'VI SUR']
df = df[df['municipio'] != 'VII SIERRA DE AMULA']
df = df[df['municipio'] != 'VIII COSTA SUR']
df = df[df['municipio'] != 'IX COSTA NORTE']
df = df[df['municipio'] != 'X SIERRA OCCIDENTAL']
df = df[df['municipio'] != 'XI VALLES']
df = df[df['municipio'] != 'XII CENTRO']
```

```python
df
```

```python
df.to_csv('data\Servicios_Turisticos_Complementarios\Servicios_Turisticos_Historico.csv', index=False)
```
