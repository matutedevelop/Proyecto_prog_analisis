# Creacion de los archivos csv de post pandemia

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Vista 02

```python
df_02 = pd.read_excel('10_1_Vistas_2-6.xlsx', sheet_name='Vista02', header=1)
df_02.rename(columns={'Unnamed: 0': 'Año'}, inplace=True)
df_02 = pd.pivot_table(df_02, [x for x in df_02.columns if x not in ['Año', 'Categoria']], ['Año', 'Categoria'])
```

```python
df_02.to_csv('alojamiento.csv')
```

## Vista 03

```python
df_03 = pd.read_excel('10_1_Vistas_2-6.xlsx', sheet_name='Vista03')
df_03.rename(columns={'Unnamed: 0': 'Año'}, inplace=True)
df_03 = df_03.pivot(index='Año', columns='Estado', values='Unnamed: 2')
df_03
```

```python
df_03.to_csv('establecimientos.csv')
```

## Vista 04

```python
df_04 = pd.read_excel('10_1_Vistas_2-6.xlsx', sheet_name='Vista04')
df_04.rename(columns={'Unnamed: 0': 'Año'}, inplace=True)
df_04 = df_04.pivot(index='Año', columns='Estado', values='Unnamed: 2')
df_04
```

```python
df_04.to_csv('cuartos.csv')
```

## Vista 05

```python
df_05 = pd.read_excel('10_1_Vistas_2-6.xlsx', sheet_name='Vista05')
df_05 = df_05.drop('Unnamed: 0', axis=1)
df_05.rename(columns={'Unnamed: 1': 'Año'}, inplace=True)
df_05 = df_05.pivot(index='Año', columns='Estado', values='Unnamed: 3')
df_05
```

```python
df_05.to_csv('establecimientos_crecimiento.csv')
```

## Vista 06

```python
df_06 = pd.read_excel('10_1_Vistas_2-6.xlsx', sheet_name='Vista06')
df_06 = df_06.drop('Año', axis=1)
df_06.rename(columns={'Unnamed: 1': 'Año'}, inplace=True)
df_06 = df_06.pivot(index='Año', columns='Estado', values='Jalisco')
df_06
```
